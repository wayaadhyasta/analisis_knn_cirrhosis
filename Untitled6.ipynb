5. Modelling

```python
import pandas as pd # Untuk manipulasi data
from sklearn.model_selection import train_test_split # Untuk membagi dataset menjadi train & test
from sklearn.neighbors import KNeighborsClassifier # Algoritma K-Nearest Neighbors (KNN)
from sklearn.metrics import accuracy_score, classification_report # Untuk evaluasi model
from sklearn.impute import SimpleImputer # Untuk menangani missing values
from sklearn.pipeline import Pipeline # Untuk membuat pipeline pemrosesan data

# Memisahkan fitur (x) dan target (y)
x = df.drop(columns=['Drug'])  # Menghapus kolom 'Drug' sebagai target prediksi
y = df['Drug'] # Menyimpan kolom 'Drug' sebagai label (target)

# Membagi dataset menjadi data latih (80%) dan data uji (20%)
# Membagi dataset menjadi data latih (train) dan data uji (test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Membuat pipeline untuk menangani missing values dan melakukan klasifikasi dengan KNN
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')), # Mengisi missing values dengan nilai yang paling sering muncul
    ('knn', KNeighborsClassifier(n_neighbors=3)) # Model K-Nearest Neighbors dengan k=3
])

# Melatih (fit) pipeline dengan data latih
pipeline.fit(x_train, y_train)

# Memprediksi data uji
y_pred = pipeline.predict(x_test)

# Menghitung akurasi model
knn_acc = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
print("Akurasi model KNN: {:.2f}%".format(knn_acc*100))
```


1. Persiapan Data:
df.drop(columns=['Drug']): Menghapus kolom 'Drug' karena itu adalah target yang ingin diprediksi.
df['Drug']: Menyimpan kolom 'Drug' sebagai label (target output).
train_test_split(): Membagi dataset menjadi 80% data latih dan 20% data uji.

2. Membuat Pipeline Pemrosesan:
SimpleImputer(strategy='most_frequent'): Mengisi missing values dengan nilai yang paling sering muncul.
KNeighborsClassifier(n_neighbors=3): Menggunakan model K-Nearest Neighbors dengan k=3.

3. Melatih Model & Prediksi:
pipeline.fit(x_train, y_train): Melatih model dengan data latih.
pipeline.predict(x_test): Memprediksi target dari data uji.

4. Evaluasi Model:
accuracy_score(y_test, y_pred): Menghitung akurasi model berdasarkan data uji.
classification_report(y_test, y_pred): Menampilkan metrik evaluasi seperti precision, recall, dan f1-score.
print("Akurasi model KNN: {:.2f}%".format(knn_acc*100)): Menampilkan akurasi dalam persen.
